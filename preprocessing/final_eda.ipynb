# Users

```python
users = pd.read_csv('./data/BX-Users.csv', encoding='latin1')
```

```python
users.columns = ['id','location','age']
null_idx = users[(users['age'] < 5) | (users['age'] > 70)].index
users.loc[null_idx, 'age'] = np.nan
users.loc[users['age'].isnull(), 'age'] = 34

users['country'] = users['location'].apply(lambda x: x.split(',')[-1])
users['region'] = users['location'].apply(lambda x: x.split(',')[-2])
users['city'] = users['location'].apply(lambda x: x.split(',')[-3])

users['country'] = users['country'].apply(lambda x: x.lstrip() if isinstance(x, str) else x)
users['region'] = users['region'].apply(lambda x: x.lstrip() if isinstance(x, str) else x)
users['city'] = users['city'].apply(lambda x: x.lstrip() if isinstance(x, str) else x)

def country_eda(x):
    if x in ["usa", "u.s.a.", "united states", "united state"]:
        return 'usa'
    elif x in ['united kingdom','england','ireland']:
        return 'united kingdom'
    elif x == 'phillipines':
        return 'philippines'
    elif x in ['far away...','here and there']:
        return 'usa'
    elif x in ['catalunya', "euskal herria", "galiza", "orense", "catalonia"]:
        return 'spain'
    else:
        return x

users['country'] = users['country'].apply(country_eda)

users.loc[users['region'].isin(['new hampshire', 'texas', 'south carolina', 'new york', 'missouri']), 'country'] = 'usa'
users.loc[users['region'] == 'nova scotia', 'country'] = 'canada'
users.loc[users['region'] == 'new south wales', 'country'] = 'australia'
users.loc[users['region'] == 'rhône alpes', 'country'] = 'france'
# 주 약칭을 풀네임으로 변환하는 딕셔너리
state_abbreviations = {
    'dc': 'district of columbia', 'ga': 'georgia', 'ga.': 'georgia', 'tx': 'texas', 
    'nm': 'new mexico', 'wa': 'washington', 'ca': 'california', 'va': 'virginia', 'co': 'colorado'
}

# 미국 실제 주 이름 리스트
us_states = [
    'alabama', 'alaska', 'arizona', 'arkansas', 'california', 'colorado', 'connecticut', 
    'delaware', 'district of columbia', 'florida', 'georgia', 'hawaii', 'idaho', 'illinois', 
    'indiana', 'iowa', 'kansas', 'kentucky', 'louisiana', 'maine', 'maryland', 'massachusetts', 
    'michigan', 'minnesota', 'mississippi', 'missouri', 'montana', 'nebraska', 'nevada', 
    'new hampshire', 'new jersey', 'new mexico', 'new york', 'north carolina', 'north dakota', 
    'ohio', 'oklahoma', 'oregon', 'pennsylvania', 'rhode island', 'south carolina', 'south dakota', 
    'tennessee', 'texas', 'utah', 'vermont', 'virginia', 'washington', 'west virginia', 'wisconsin', 
    'wyoming'
]

# 주 이름을 약칭에서 풀네임으로 변환
users['region'] = users['region'].replace(state_abbreviations)

# 주 이름이 아닌 값은 NaN으로 처리
users['region'] = users['region'].apply(lambda x: x if x in us_states else np.nan)

```
